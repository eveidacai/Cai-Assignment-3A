````python
import csv
import re
````

````python
#firstname
reader = csv.reader(open('recweekdata.csv','r'))

first_name_list = []
for row in reader:
    first_name_list.append(row[0])
````

````python
#lastname
reader = csv.reader(open('recweekdata.csv','r'))

last_name_list = []
for row in reader:
    last_name_list.append(row[1])
````

````python
#birthmonth
reader = csv.reader(open('recweekdata.csv','r'))

dirty_month_list = []
for row in reader:
    dirty_month_list.append(row[2])
    clean_month_list = [re.sub(r'(January|1|Jan)',r'January',month) for month in dirty_month_list]
    clean_month_list = [re.sub(r'(February|2|Feb)',r'February',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(March|3|Mar)',r'March',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(April|4|Apr)',r'April',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(May|5|May)',r'May',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(June|6|Jun)',r'June',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(July|7|Jul)',r'July',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(August|8|Aug)',r'August',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(September|9|Sep)',r'September',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(October|10|Oct|January0)',r'October',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(November|11|Nov|JanuaryJanuary)',r'November',month) for month in clean_month_list]
    clean_month_list = [re.sub(r'(December|12|Dec|JanuaryFebruary)',r'December',month) for month in clean_month_list]
````

````python
#birthday
reader = csv.reader(open('recweekdata.csv','r'))

day_list = []
for row in reader:
    day_list.append(row[3])  
````

````python
#birthyear
reader = csv.reader(open('recweekdata.csv','r'))

dirty_year_list = []
for row in reader:
    dirty_year_list.append(row[4])
    clean_year_list = [re.sub(r'^97',r'1997',year) for year in dirty_year_list]
    clean_year_list = [re.sub(r'^98',r'1998',year) for year in clean_year_list]
    clean_year_list = [re.sub(r'^99',r'1999',year) for year in clean_year_list]
    clean_year_list = [re.sub(r'^00',r'2000',year) for year in clean_year_list]
    clean_year_list = [re.sub(r'^01',r'2001',year) for year in clean_year_list]   
````

````python
#course: does not clean BS Management properly
reader = csv.reader(open('recweekdata.csv','r'))

dirty_course_list = []
for row in reader:
    dirty_course_list.append(row[5])
    raw_clean_course_list = [re.sub(r'(BS LM|Legal Management|LM)',r'BS Legal Management',course) for course in dirty_course_list]
    raw_clean_course_list = [re.sub(r'(BS ME|Management Engineering|ME)',r'BS Management Engineering',course) for course in raw_clean_course_list]
    raw_clean_course_list = [re.sub(r'(BS ITE|Information Technology Entrepreneurship|ITE)',r'BS IT Entrepreneurship',course) for course in raw_clean_course_list]
    raw_clean_course_list = [re.sub(r'(BS CTM|Communications Technology|CTM)',r'BS Communications Technology',course) for course in raw_clean_course_list]
````

````python
# course: cleaning BS Management (manual intervention)
unique_course = set(raw_clean_course_list)
unique_course_dictionary = {course: None for course in unique_course}

for i in unique_course_dictionary:
    final_course = input(i+": "+"Enter course: ")
    unique_course_dictionary[i]=final_course
````

````python
#course: final clean course list
clean_course_list = [re.sub(r''.join(course),r''.join(unique_course_dictionary[course]),course) for course in raw_clean_course_list]
````

````python
#phonenumber
reader = csv.reader(open('recweekdata.csv','r'))

dirty_number_list = []
for row in reader:
    dirty_number_list.append(row[6])
    clean_number_list = [re.sub(r'\+63',r'',number) for number in dirty_number_list]
    clean_number_list = [re.sub(r'9',r'+63 9',number,1) for number in clean_number_list]
````

````python
#converting to CSV file
import csv

with open('cleanrecweekdata.csv','w+', newline = "") as clean_data:
    csv_writer = csv.writer(clean_data)
    
    for i in range(len(first_name_list)):
        csv_writer.writerow([first_name_list[i],last_name_list[i],clean_month_list[i],day_list[i],clean_year_list[i],clean_course_list[i],clean_number_list[i]])
````
